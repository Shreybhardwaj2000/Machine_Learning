# Decision Tree Classification Model

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('salaries.csv')
df
```

```python
inputs=df.drop('salary_more_then_100k', axis='columns')
target=df['salary_more_then_100k']
```

```python
inputs
```

```python
target
```

```python
# Now lets convert our objects into int
from sklearn.preprocessing import LabelEncoder
```

```python
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
inputs['company_n']= le_company.fit_transform(inputs['company'])
inputs['job_n']= le_company.fit_transform(inputs['job'])
inputs['degree_n']= le_company.fit_transform(inputs['degree'])
```

```python
inputs.head()
```

```python
inputs_n=inputs.drop(['company','job','degree'], axis='columns')
inputs_n
```

```python
from sklearn.model_selection import train_test_split
```

```python
inputs_n_train, inputs_n_test, target_train, target_test= train_test_split(inputs_n, target, test_size=0.1)
```

```python
inputs_n_test
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n_train, target_train)
```

```python
model.score(inputs_n_test, target_test)
```

```python
model.predict(inputs_n_test)
```

```python
target_test
```
